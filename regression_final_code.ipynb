```python
# Import necessary libraries for data manipulation and visualization
import pandas as pd  # Pandas for data manipulation and analysis
import seaborn as sns  # Seaborn for statistical data visualization
import matplotlib.pyplot as plt  # Matplotlib for creating static, interactive, and animated visualizations in Python

# Compute the correlation matrix for the 'spotify' DataFrame
# The 'corr()' method calculates the pairwise correlation of all columns in the DataFrame (excluding NA/null values)
# Correlation values range from -1 to 1, where 1 means a perfect positive linear relationship, -1 means a perfect negative linear relationship, and 0 means no linear relationship
corr_matrix = spotify.corr()

# Set up the matplotlib figure with a specific size to ensure the heatmap is clearly visible
plt.figure(figsize=(10, 8))

# Draw the heatmap using seaborn
# 'annot=True' displays the correlation values on the heatmap
# 'fmt=".2f"' formats the annotation to 2 decimal places
# 'cmap='coolwarm'' sets the color map to 'coolwarm', which is good for highlighting differences in correlation values
# 'square=True' ensures that each cell of the heatmap is square-shaped
# 'cbar_kws={"shrink": .5}' modifies the color bar size to fit the figure aesthetically
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .5})

# Adjust layout to make room for the rotated x-axis and y-axis labels
# 'rotation=45' rotates labels to 45 degrees to avoid overlap and improve readability
# 'ha='right'' aligns the x-axis labels to the right for a better layout
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=45)

# Adjust the layout of the plotting area to fit the rotated labels and ensure the plot is neatly presented
plt.tight_layout()

# Display the heatmap plot
# This visualization helps in understanding how different features in the 'spotify' DataFrame correlate with each other
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import math



X_train= spotify.drop(columns=['Id','title','artist','popularity','spch','val','live','top genre'])
y_train=  spotify['popularity']


std_scaler = preprocessing.StandardScaler()
X_train = std_scaler.fit_transform(X_train)

# 1. Define the model with chosen parameters
model = GradientBoostingRegressor(n_estimators=200, learning_rate=0.01, max_depth=3, random_state=42)

# 2. Train the model
model.fit(X_train, y_train)












```

```python
# 3.Load the test data
test_df=pd.read_csv("CS98XRegressionTest.csv")
test_ids=test_df["Id"]
```

```python
# 4.Prepare our data by selecting only the features that we are interested in

X_test= test_df.drop(columns=['Id','title','artist','val','live','spch','top genre'])

# 5.Scale features to standardize them for better performance of the model

std_scaler = preprocessing.StandardScaler()
X_test = std_scaler.fit_transform(X_test)


submission_pred=model.predict(X_test)

#Prepare the submission file
df= pd.DataFrame({"Id":test_ids.values,"pop":submission_pred})
df.to_csv("submission.csv",index=False)


```
